# Flight Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
# Assuming train_df and test_df are your dataframes
final_df = pd.concat([train_df, test_df], ignore_index=True)
final_df.tail()
```

# Feautre engineering 

```python
final_df['Date_of_Journey'].str.split('/').str[0].unique
```

```python
final_df["Date"]=final_df['Date_of_Journey'].str.split('/').str[0]
final_df["Month"]= final_df['Date_of_Journey'].str.split('/').str[1]
final_df["Year"]=final_df['Date_of_Journey'].str.split('/').str[2]  
```

```python
final_df.head(2)
```

```python
final_df["Date"]= final_df["Date"].astype(int)
final_df["Month"]=final_df["Month"].astype(int)
final_df["Year"] = final_df["Year"].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis= 1, inplace= True)
```

```python
final_df.head(2)
```

```python
final_df["Arrival_Time"]=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head(1)
```

```python
final_df.isnull().sum()
```

```python
final_df["Arrival_hours"]=final_df['Arrival_Time'].str.split(':').str[0]
```

```python
final_df["Arrival_hours"]= final_df["Arrival_hours"].astype(int)
```

```python
final_df["Arrival_Min"]=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df["Arrival_Min"]=final_df['Arrival_Min'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0 , '2 stops' :2, '1 stop':1, '3 stops' :3, 'nan':5, '4 stops':4})  
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df.drop('Route',inplace=True, axis=1)
```
